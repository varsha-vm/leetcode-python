```python
#sort the list by length
fruits_list = ["apple", "kiwi", "banana"]
key = lambda x: len(x)
sorted(fruits_list, key=key)
```

```python
#Sort list of tuples by second element
tuples = [(1, 9), (2, 3), (4, 1)]
key=lambda x: x[1]
sorted(tuples, key=key)
```

```python
#Ô∏èsort dictionary by key
hashmap ={"b":2, "c":1, "a":3}
key = lambda x: x[0]
sorted(hashmap.items(), key=key)
```

```python
#Sort dictionary by value
key = lambda x: x[1]
sorted(hashmap.items(), key=key)
```

```python
#Sort words by last letter
list_of_words = ["sun", "apple", "car"]
key = lambda x : x[-1]
sorted(list_of_words, key=key)
```

```python
#Sort tuples by sum of elements
tuple_nums = [(1, 2), (10, 1), (3, 3)]
key=lambda x: x[0] + x[1]

sorted(tuple_nums, key=key)
```

```python
#Sort list of dicts by age
people = [
    {"name":"Alice", "age":30},
    {"name":"Bob", "age":20},
    {"name":"Eve", "age":25}
]

key = lambda x: x["age"]

sorted(people, key=key)
```

```python
#Sort strings by number of vowels ('a','e','i','o','u')
words = [ "hello", "sky", "aaa"]
key = lambda x: sum( c in "aeiou" for c in x)
sorted(words, key=key, reverse=True)
```

```python
#Sort by absolute value

value = [-10, 1, -2, 5, -7]
key = lambda x: abs(x)
sorted(value, key=key)
```

```python
#Sort by frequency
from collections import Counter

nums = [1,1,2,3,3,3]
freq = Counter(nums)


key = lambda x: x[1]
sorted(freq.items(), key=key)
```

```python
#Sort by multiple keys (first value desc, second asc)
values = [(1,2), (1,1), (2,2), (2,1)]

key = lambda x: (-x[0], x[1])
sorted(values, key=key)
```

```python
#Sort intervals by start, then end
iterable_lst = [[1,3], [1,2], [2,4]]
key = lambda x: (x[0],x[1])
sorted(iterable_lst, key=key)
```

```python
#Sort list of strings in reverse alphabetical by their second letter
string_lst = ["cat", "dog", "apple"]
key = lambda x: x[1]
sorted(string_lst, key=key, reverse = True)
```

```python
#Sort matrix rows by the sum of each row
nested_lst = [[10], [1,2,3], [5,1]]
key = lambda x: sum(x)
sorted(nested_lst, key=key)
```

```python
#Sort objects by computed score
students = [
    {"name":"A", "math":90, "eng":70},
    {"name":"B", "math":80, "eng":95},
]

key = lambda x: (x["math"]*0.7 + x["eng"]*0.3)

sorted(students, key=key, reverse=True)
```
